```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# True model parameters
a0 = 5
a1 = 10
sig = 5

# Create dataset
N = 10 #number of samples
rng = np.random.default_rng()
x = rng.uniform(0, 10, N)
y = np.zeros((len(x),1))
for i in range(len(x)):
    y[i] = rng.normal(a0+a1*x[i], sig,1)
```

```python
# Generate x values for plotting
x_start = min(x)
x_stop = max(x)
dx = (x_stop - x_start)/1000
xplot = np.arange(min(x), max(x), dx)
```

```python
# Linear regression via normal equations for model parameters
A = np.matrix([x**0, x**1]).T

model_params = np.linalg.inv(A.T.dot(A)).dot(A.T.dot(y))

model_a0 = model_params[0,0]
model_a1 = model_params[1,0]

# Predictions - true model for plotting
linear_model = model_a0 + model_a1*xplot
linear_ground_truth = a0 + a1*xplot
```

```python
# Plot results
colors = np.ones(N)
area = (5 * np.ones(N))**2  
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.plot(xplot, linear_ground_truth, label = "True Distribution")
plt.plot(xplot, linear_model, label = "Model Distribution")
plt.legend()
plt.show()
```
