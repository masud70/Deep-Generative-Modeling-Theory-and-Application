```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# True model parameters
mu = 1
sig = 2

# Number of samples to draw from the true model
N = 100

# Create dataset
rng = np.random.default_rng()
x = rng.normal(mu, sig, N)
```

```python
# Create a plottable version of the true model
x_start = min(x)
x_stop = max(x)
dx = (x_stop - x_start)/1000
xplot = np.arange(min(x), max(x), dx)
gaussian = 1/math.sqrt(2*math.pi)*sig*np.exp(-0.5*(xplot-mu)**2/sig**2)
```

```python
# Predictions - Model parameters come from MLE equations
y = np.zeros(np.shape(x))
model_mu = 1/N*np.sum(x)
model_sig = math.sqrt(1/N*np.sum((x - model_mu)**2))
gaussian_model = 1/math.sqrt(2*math.pi)*model_sig*np.exp(-0.5*(xplot-model_mu)**2/model_sig**2)
```

```python
# Plot results
colors = np.ones(N)
area = (5 * np.ones(N))**2  #radius of scatter circle
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.plot(xplot, gaussian, label="True Distribution")
plt.plot(xplot, gaussian_model, label="Model Distribution")
plt.legend()
plt.show()
```
